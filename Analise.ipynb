```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_excel("Headphones_trat.xlsx")
```

```python
Cores = df["Cor"].value_counts()
plt.figure(figsize=(10,5))
Cores.plot(kind="bar", color="blue")
plt.title("Cores mais Vendidas")
plt.xlabel("Cores")
plt.xticks(rotation=45)
plt.show()
```

```python
# A Kabum tem preferencia em anunciar produtos com cores neutras e não chamativas
```

```python
possui_microfone = df["Microfone"].value_counts()
plt.figure(figsize=(10,5))
wedges, texts = plt.pie(possui_microfone, labels=None, startangle=90)

plt.legend(wedges, possui_microfone.index, loc="center left", bbox_to_anchor=(1, 1))
plt.title("Tem microfone?")
plt.tight_layout()
plt.show()
```

```python
# Grande parte dos headphones anuniados são com o produto desacompanhado de microfones
# Poderia haver mais vendas se fosse investido em mais fones com microfones
```

```python
Conexao = df["Conexão"].value_counts()
plt.figure(figsize=(10, 5))
wedges, texts = plt.pie(Conexao, labels=Conexao.index, startangle=90,)
plt.legend(wedges, Conexao.index, loc="center left", bbox_to_anchor=(1, 1))
plt.title("Tipo de conexão")
plt.tight_layout()
plt.show()
```

```python
# Um pouco acima da metade dos headphones anunciados são bluetooth
# Ou seja, se forem anunciados mais fones bluetooth, pode haver um aumento nas vendas,
# por expor produtos que mostram organização e simplicidade, ao invés de apresentar fios
```

```python
Media_Valor = df["Preço(num)"].mean()
print(f"Média dos preços dos headphones: R$ {Media_Valor:.2f}")
```

```python
# O valor médio de todos os headphones anunciados é R$427,50
```

```python
desvio_padrao_preco = df["Preço(num)"].std()
print(f"Desvio padrão dos preços dos headphones: R$ {desvio_padrao_preco:.2f}")
```

```python
# O desvio padrão dos valores de todos os headphones anunciados é R$823,49
```
