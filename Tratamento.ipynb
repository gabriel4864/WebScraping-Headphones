```python
import pandas as pd
import numpy as np

df = pd.read_excel("headphones.xlsx")
display(df)
```

```python
df['Preço(num)'] = (
    df['preco']
    .str.replace('R$', '', regex=False)      
    .str.replace('.', '', regex=False)       
    .str.replace(',', '.', regex=False)      
    .astype(float)                           
)

# Exibe o DataFrame
display(df)
```

```python
cores = [
"amarelo", "amarela", "azul", "branco", "branca", "cinza", "creme", "dourado", "laranja", "marrom", "preto", "rosa", "roxo", "verde", "Vermelho"
"bege", "graphite", "geada", "uva", "grey", "yellow", "blue", "silver", "red", "black", "off white", "white", "green"
]

def extrair_cor(texto):
    texto = texto.lower()
    for cor in cores:
        if cor in texto:
            return cor
    return "Desconhecida"

df["cor_principal"] = df["marca"].apply(extrair_cor)
display(df[["marca", "cor_principal"]])   
```

```python
contagem = df['cor_principal'].str.contains('Desconhecida', case=False, na=False).sum()

print(f'A palavra "desconhecida" aparece {contagem} vezes na coluna.')
```

```python
df['cor_principal'] = df['cor_principal'].str.title()
display(df)
```

```python
df_filtrado = df[df['cor_principal'].str.contains('desconhecida', case=False, na=False)]

display(df_filtrado)
```

```python
df['Conexão'] = df['marca'].str.contains('bluetooth|sem fio|s/ fio', case=False, na=False).replace({
    True: 'Bluetooth',
    False: 'Com fio'
})

display(df)
```

```python
df['Microfone'] = df['marca'].str.contains('microfone', case=False, na=False).replace({
    True: 'Sim',
    False: 'Não'
})

display(df)
```

```python
df['Faixa Etária'] = df['marca'].str.contains('infantil|criança|kids', case=False, na=False).replace({
    True: 'Infantil',
    False: 'Geral'
})

display(df)
```

```python
df = df.rename(columns={
    'marca': 'Produto',
    'preco': 'Valor',
    'cor_principal': 'Cor'
})
```
